### Pandas Series is a column of data in a table like in excel or Google Sheets with labels called as indexes attached to each value

```python
pd.Series([40,50,60], index=['Bread','Milk','Butter'], name='Price of Grocery Items')
```

```python
pd.Series(data = [40,50,60], index=['Bread','Milk','Butter'], name='Price of Grocery Items')
```

```python
## Series is a 1D array that can hold any type of data like integers, float, strings(object),
#  boolean, mixed data type.
```

#### Data type --> Integers --> Data in a Series(Column) --> Quantity of items in the stock 
#### Data type --> Float --> Data in a series(Column) --> Price or product ratings
#### Data type --> Strings --> Column --> Product Name
#### Data type --> Boolean --> Column --> Is item in stock?
#### Data type --> Dates/times --> Column --> expiry dates or sales dates
#### Mixed Types --> Milk, 40, True in a single column 

### Creating a Series in Multiple Ways:
1. List
2. NumPy Array
3. Dictionary
4. Custom Index
5. Scalar value

```python
### Series from list

country = ['India', 'China', 'USA', 'Pakistan']
pd.Series(country)
```

```python
batsman_runs = [48,64,72,98,114,183]
pd.Series(batsman_runs)
```

```python
batsman_runs = [48,64,72,98,114,183]
pd.Series(batsman_runs, dtype = 'int32')
```

```python
subject_names = ['Maths','Physics','Chemistry','Biology','English']
subject_marks = [48, 92, 64, 35, 98]

pd.Series(subject_marks, index = subject_names, name = 'Subject Marks Data', dtype = 'int32')
```

```python
## Creating a Series using NumPy Array
values = np.arange(10,21)
pd.Series(values)
```

```python
## Dictionary to Series
## Dictionary is another Data Structure that stores the data in a key value pair and just because 
# it stores the data in a key value pair so we don't have to mention indexes explicitly 
```

```python
data_items = {'Maths' : 78, 'English' : 92, 'Physics' : 88, 'Chemistry': 64}
pd.Series(data_items)
```

```python
## custom indexes to series

values = [1,2,3,4,5]
pd.Series(data = values, index = ['a','b','c','d','e'])
```

```python
## Scalar value inside a series

pd.Series(4, index=['a','b','c','d','e'])
```

## Series Methods

#### 1. head() --> Provides the top preview of data as the top 5 rows of the dataframe or Series. It is quite helpful with providing the preview of the dataset and if we want to preview top `10 rows, then we can pass 10 as argument, for top 7 rows, pass top 7 rpws as an argument.

#### 2. tail() --> provides the bottom preview of the data as the bottom 5 rows.

#### for eg: dataset ---> salary column ---> top 5 rows ---> salary --> 7 lpa ---> top 10 --> top 15,,,  tail()---> bottom 5 rows --> salary --> 4 lpa, 9 lpa  

#### 3. sample()  ---> sample() provides any custom row from the dataset for us to preview it

#### value_counts() ---> it provides the idea about the count of the values(categorical values). if i am havng a column where i want to calculate categorical value count then i will go with value_counts(). 

#### sort_values() ---> this method sorts the values in an ascending manner. It can be useful for us for the numerical columns. you can even sort the values in descending manner as well.

#### sort_index() ---> this method sorts the index of the dataset
