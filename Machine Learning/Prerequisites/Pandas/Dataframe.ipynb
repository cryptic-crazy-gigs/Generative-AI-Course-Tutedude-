## Functions

#### 1. head() --> returns the top 5 rows of the dataframe. when given a number as a parameter: returns that number of top rows from the dataframe

#### 2. tail() --> returns the bottom 5 rows of the dataframe. when given a number as a parameter: return that number of bottom rows from the dataframe

#### 3. sample() --> return random 1 row from the dataset. if given a number as a parameter, returns that many random rows from the dataset. Grt way to understand the variation in dataset

#### 4. info() --> provides  the idea --->  

#### 1. about the datatype of all columns
#### 2. Number of Missing values of all columns
#### 3. DataFrame storage in memory

#### 5. describe() --> it gives the mathematical summary of the data -> min, max, mean, std, count(how many values are there), 25%ile, 50%ile, 75%ile
