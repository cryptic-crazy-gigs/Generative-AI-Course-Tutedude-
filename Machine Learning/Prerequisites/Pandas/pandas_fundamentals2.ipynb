### Mathematical Functions in Pandas

#### 1. sum() ---> sum of all values present in the series
#### 2. product() ---> product of all values present in the series
#### 3. mean() ---> provides us with the averge value of the total column
#### for eg---> Company Data---> Salary ---> average value for that column salary = 7,6,8,9,11 mean

#### 4. median() ---> median provides the idea about the average of the dataa. to reach median, we need to sort the values, the middle valuee is the median.
#### data --> salary ---> 1 crore, 10 crore (outlier) --> avg -> 7+8+6+9
#### outlier ---> that value which is either too big or too small to be called a part of a dataset. such values resut in largee difference between the average and median.

#### mode() --> Mode provides us with the idea that which value is getting repeated maximum number of times.


#### std() ---> provides the idea that how the values inside the series are distributed from the average or mean value - standard deviation
#### salary ---> 7,8,6,6.5,7.5 ----> low standard deviation
#### salary ---> 7,15,4,20,28,17 ----> high standard deviation


#### var() ---> variance also provides the insight about the distribution of the data from the mean value and varance is the square of std.
#### vaariance = (mean - v1)**2 + (mean - v2)**2 +(mean - v3)**2  + .... /(total count of values)

#### variance = std**2

#### min() ---> minimum value in the series
#### max() ---> maximum value in the series

#### describe() ----> describe fn provides the idea about the summary of the important mathematical calculations for the data

#### describe function has :
    min()
    max()
    mean()
    std()
    25%---> 25% of the values out of all the values are less than this specific value
    50%---> 50% of the values out of all the values are less than this specific value
    75%---> 75% of the values out of all the values are less than this specific value
    count() --> total number of rows in the column
