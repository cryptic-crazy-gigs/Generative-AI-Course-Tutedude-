#### NumPy is a fundamental package for scientific computing in python. It is a Python Library that provides a multi Dimensional Array object and various derived objects like masked arrays or matrices and an assortment of routines for fast operations on Arrays including Mathematical, Logical, Shape Manipulation, Sorting, Selecting, I/O, Discrete Fourier Transform, Basic Linear Algebra, Statistical Operations, Random Simulation, etc.

#### At the core of NumPy package is the N-dimensional Array object

## NumPy Arrays vs Python Sequences

#### NumPy Arrays have a fixed size at creation, unlike Python lists which grow dynamically. Changing the size of an n-dimensional array will create a new array and delete the original

#### The elements in a numpy array are required to be of the same data type, and hence each element will occupy same size in memory

#### NumPy Arrays facilitate advanced Mathemetical and other types of operations on large number of data. Typically, such operations are executed more efficiently with less code than is possible using Python's built-in sequences
