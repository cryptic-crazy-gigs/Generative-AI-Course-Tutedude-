# Stacking and Splitting in Array

```python
## Stacking --> joining the arrays together
## Horizontal Stacking --> Row by Row joining is done --> hstack()
## Vertical Stacking --> column by column joining is done --> vstack(a)
```

```python
array_one = np.arange(1,13).reshape(3,4)
```

```python
array_one
```

```python
array_two = np.arange(13,25).reshape(3,4)
```

```python
array_two
```

```python
# lets do horizontal stack
horizontal_array = np.hstack((array_one, array_two))
```

```python
horizontal_array
```

```python
horizontal_array = np.hstack((array_one, array_two, array_one, array_two))
```

```python
horizontal_array
```

```python
vertical_array = np.vstack((array_one, array_two))
```

```python
vertical_array
```

```python
## Horizontal Split --> np.hsplit(array, no. of parts you want your array to be splitted at)
## Vertical Split --> np.vsplit(array, no. of parts you want your array to be splitted at)
```

```python
np.hsplit(vertical_array, 4)
```

```python
np.hsplit(vertical_array, 2) # how many splits at parameter 2
```

```python
np.vsplit(vertical_array,2)
```

```python
ar1, ar2, ar3 = np.vsplit(vertical_array,3)
```

```python
ar1
```

```python
ar2
```

```python
ar3
```

```python
for i in np.vsplit(vertical_array,3):
    print(i)
    print("\n")
```
