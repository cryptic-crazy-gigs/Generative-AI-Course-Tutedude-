## Stacked Bar Chart

#### Instead of keeping the bars in side by side direction, we can keep the bars in the stacked manner i.e, One on top of other and in order to figure out which category is having Maximum Sale, we can have a look over the length of the bar

## Stacked Bar Chart ---> bar() --> parameter--> bottom --> This bottom parameter speciifies what bar i want to keep below my current bar

#### Red Bar
#### Green Bar ---> Bottom ---> Red Bar
#### Blue Bar ----> Bottom --> Green Bar + Red Bar

### bottom = place from where you want to start this bar. default = 0; it is the height from which you want to start the bar

```python
plt.figure(figsize=(7,4))
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.bar(m.index, m.values, color = 'green', label='Mint Chip Choco')
plt.bar(d.index, d.values, color = 'orange', label='85% Dark Bars',bottom=m.values)
plt.bar(p.index, p.values, color = 'pink', label='Peanut Butter Cubes',bottom=m.values+d.values)
plt.legend()
plt.show()
```
