```python
den = int(input("Enter denominator :"))

try:
    result = 1/den
    print(result)
except ZeroDivisionError as ex:
    print(ex)
```

```python
b = input("enter denominator :")

try:
    result = 1/b
    print(result)
    a=c
except ZeroDivisionError as ex1:
    print(ex1)

except NameError as ex2:
    print(ex2)

except Exception as ex:    # when we are not aware of what kind of error can happen, we can just write this way
    print(ex)              # no need to know which kind of error class it is 
```

### If we are getting an error and we are not aware that which specific type of error it is then we can use the primary exception class which is Exception

### The parent exception type must be mentioned at the end or at the last as it is the parent class so only if the previous exception class will not catch the error then this final one will be able to catch it.

```python
# value error
try :
    a = int("123")
    print(a)
    b = int("abc")
    print(b)
except ValueError as ex:
    print(ex)
```

```python
### try ----> try block is going to have the operation where we can expect an error to occur
### except ----> which is used to handle the error in a way such that the program execution should not stop

### else ---> It will be consisting of the code which we want to run if the code inside try block runs without throwing an error
### finally ---> This block runs after we are done with everything above. It does not matter if the try block threw an error or not
```

#### if try ---> code is throwing an error ---> except ----> code will be running  (Except)
#### if try ---> code is not throwing an error(working fine) --->ELse ---> code will be running (Else)

```python
num = int(input ("enter denominator :"))

try :
    result = 1/num
except Exception as ex :
    print(ex)
else:
    print("The division result for your denominator value will be ", result)
```
