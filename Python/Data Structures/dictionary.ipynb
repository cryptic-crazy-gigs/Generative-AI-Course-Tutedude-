so, let's talk about copies.

shallow copy: shallow copy is where if there are changes in the original copy, they are not reflected in the shallow copy.  .copy() for this

deep copy: it is where the changes in one are also reflected in nother. It is truly not a copy. It is just the same thing has 2 names now and so it can be accessed by 2 names.

```python
dict_1 = {"Name":"Garvi", "Age":26, "greeting":"Hi"}

dict_2 = dict_1  #dict_2 is deep copy
```

```python
dict_3 = dict_1.copy()
```

```python
dict_1["Name"] = "garvita"
```

```python
print(dict_1)
print(dict_2)
print(dict_3)
```

Nested Dictionary:
for keeping dictionary inside another dictionary

```python
## to store the information for multiple students : name, age, grade

nested_dict = {"Student_1":{"Name":"Amit", "Marks":84,"Grade":"A"},
               "Student_2":{"Name":"Vivek", "Marks":93,"Grade": "A+"},
               "Student_3":{"Name":"Vikas", "Marks":98,"Grade":"A+"}
               }

```

```python
nested_dict
```

```python
for outer_key, outer_val in nested_dict.items():
    print(f"The outer key is {outer_key} to which the corresponding  dictionary is {outer_val}")
    for inner_key, inner_val in outer_val.items():
        print(f"Inner key: {inner_key}, Inner val: {inner_val}")
```

## dictionary comprehension
#### creating a dictionary by using looping statements, and conditional statements inside the dictionary

{expression for item in iterable if condition}

```python
comprehensive = {x**2 for x in range(11)}
```

```python
comprehensive  # a set
```

```python
comprehensive = {x:x**2 for x in range(11)}
```

```python
comprehensive
```

```python
even_dictionary = {x:x**2 for x in range(1,11) if x%2==0}
```

```python
even_dictionary
```

```python
dict_1 = {"a":1, "b":2, "c":3}
dict_2 = {"b":4, "c":9, "d":16}

merged_dict = {**dict_1, **dict_2}
merged_dict
```
