## List Methods

#### 1. append -> add element at end of list (element)
#### 2. insert -> add element at specific index (index, element)
#### 3. remove -> remove the 1st occurence of the element. parameter:element
#### 4. pop -> remove the elemt at specified index. parameter(index), return the element at that index as well
#### 5. index -> returns the index of the parameter element
#### 6. reverse -> reverses the elements of the whole list. the change is seen in the original list.
#### 7. count -> returns the count of the parameter element in the list.
#### 8. sort -> sorts the original list 
#### 9. clear -> the list becomes empty on its use.

### we are going to use looping statements inside the list to store the values inside the list in an iterable format

### syntax:
[expression for item in iterable if condition]

expression : what to be printed
for item in iterable : looping  statement
if condition : conditional statement

```python
list_items = [x for x in range(1,51)]
list_items
```

```python
numbers_squared = [item for item in range(1,16)]
numbers_squared
```

```python
even_items = [x**2 for x in range(1, 19) if x%2 == 0]
even_items
```

```python
list_itemss = ["Amit", "Apple", "Banana", "car", "truck", "laptop", "technology"]
length_list = [len(word) for word in list_itemss]
length_list
```

## Nested List Comprehension
#### we will be having a combination of multiple lists in a single list itself

```python
pair = [[i,j]for i in list_itemss for j in length_list]
pair
```

```python
lists = [[list_itemss[i], length_list[i]]for i in range(0, len(list_itemss))]
lists
```

```python
lst1 = [1, 3, 4]
lst2 = [2, 6, 9]

paired_list = [[i,j]for i in lst1 for j in lst2]
paired_list
```
