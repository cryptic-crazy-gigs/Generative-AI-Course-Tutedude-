### 2 tuple methods
#### 1. count() 
#### 2. index()

```python
tuple_methods = (1, 2, 3, 8, 6, 9, 4, 2, 3, 1, 2, 2)
print(tuple_methods.index(2))
print(tuple_methods.count(2))
```

```python
#tuples inside neste list
tuple_list = ((1,2,3),(4,5,6),(11,12,13,324))
tuple_list[2][:]
```

```python
#nested tuple
Nested_tuple = ((1,2,3),(4,5,6,7),(13,15,17,32))
Nested_tuple[0][:]
```

```python
for i in Nested_tuple:
    for j in i:
        print(j, end =" ")
    print(end ="\n")
```
