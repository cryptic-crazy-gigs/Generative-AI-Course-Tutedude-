## File Handling

#### 1. Read the content from my text file in the system


```python
## open() ---> opening your file
## first parameter  --> file name
## second parameter --> mode
## mode --> It decides the way in which we are working with this file

'''
Mode -> r ----> Reading from file
Mode -> w ----> Writing into file
Mode -> r+ ---> First reading, then writing
Mode -> w+ ---> First writing, then reading
'''
```

```python
## Reading your file
## read() --> reads the content from the file
```

```python
f = open('example.txt', 'r')  # opened the file and put into r
content = f.read()
print(content)
f.close()
```

### Writing the content into the file

### write() ---> it overwrites the content into the text file
### mode -> w -> content will be overwritten
### mode -> a -> content will be added to the end of file

```python
## write() ---> helps us out in writing the content into the file
```

```python
f = open('example.txt', 'w')
f.write('Hello, We are now writing the content into the file')
f.close()
```

```python
f = open('example.txt', 'a')
f.write('\nThis is the second line of the page')
f.close()
```

### with ---> it will be used for first of all opening the file as an object and there is no need for closing the file explicitly



```python
with open('example.txt','r') as f:
    content_data = f.read()
    print(content_data)
```

```python
with open('example.txt', 'w') as file :
    file.write('We are exploring with command in python file handling')
```

```python
with open('example.txt', 'a') as files :
    files.write('\nWe are exploring appen mode as well')
```

```python
## writelines()  ----> write the content into our file line by line
```

```python
lines = ['First Line\n', 'Second Line\n', 'Third Line\n']

with open('example.txt','w') as data :
    data.writelines(lines)

# quite similar to write. writing things from the end of line.
# just writelines() ----> parameter is a list. writes elements of it 
# write() ----> takes a string as a parameter
```

```python
## mode --> Write the content into the binary file
## mode --> Read the content from the binary file
```

```python
data = b'\x00\x01\x02\x03\x04'

with open('example2.txt', 'wb') as binary_file :
    binary_file.write(data)
```

```python
with open('example2.txt', 'rb') as binary_file :
    binary_content = binary_file.read()
    print(binary_content)
```

```python
# read the content from the source file and then writing the content into another file
```

```python
with open('source.txt','r') as source :
    content = source.read()

with open('destination.txt','w') as destination:
    destination.write(content)
```

```python
## Reading the total lines, total word, total characters -> readlines()

def count_lines(file):
    with open('total_data.txt','r') as files:
        lines = files.readlines()
        total_lines = len(lines)
        total_words = sum (len(line.split()) for line in lines)
        total_characters = sum(len(line) for line in lines)
    return total_lines, total_words, total_characters

file_path  = 'total_data.txt'
total_lines, total_words, total_characters=count_lines(file_path)
print(f"total lines are {total_lines}, total words are {total_words}, total characters are {total_characters}")
```

w+ ---> First of all as W is there so it states that first writing operation will be done and then reading operation will be done
seek() --> it will help us out with reading the content after writing from the start only

```python
with open('total_data.txt','w+') as write_read_file :
    write_read_file.write('\nwe are looking into w+ operation')
    write_read_file.write('\nIt means writing aand then reading')
    content_part = write_read_file.read()
    print(content_part ," hello")
    write_read_file.seek(0)
    content_total = write_read_file.read()
    print(content_total)
```
