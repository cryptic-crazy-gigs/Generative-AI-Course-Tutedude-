```python
## os.path.isfile()  ---->  This checks if the current path is a file or not
## os.path.isdir()   ---->  This checks if the current path is a directory or not

path = "total_data.txt"

if os.path.isfile(path) :
    print("The current path belongs to a file")
elif os.path.isdir(path) :
    print("The current path belongs to a directory")
else :
    print("current path is neither a file nor a directory")
```
