```python
class Avenger:

    def __init__(self,name,power,weapon,nationality = "American"):
        self.name = name
        self.power = power
        self.weapon = weapon
        self.nationality = nationality

    def introduction(self):
        print(f"Hello, My name is {self.name}")

    def carrying_weapon(self):
        print(f"I am carrying the weapon {self.weapon}")
```

```python
Iron_man = Avenger("Iron Man", "Genius Intellect and Technology", "Repulsor Beams")

Iron_man.introduction()
Iron_man.carrying_weapon()
```

## Type of Constructors :
### 1. Default Constructor
### 2. Parameterized Constructor

### Default Constructor : It does not have the parameters inside it and it is only having the self keyword inside it. It assigns a default value to all of the objects.

### Parameterized Constructor : It consists of parameters inside it so that every object can have a specific attribute associated to it. 
### It allows us to pass arguments while creating an object

```python
class Dog :

    def __init__(self):
        self.name = "Labrador"
        self.color = "brown"

    def show_info(self):
        print(f"This dog is a {self.name} and the color of the dog is {self.color}")
```

### Types of Method :

#### 1. Instance Method : Instance Method operate on object attributes.
#### The Instance Method is going to use self keyword to access and modify object attributes.
#### These methods are called using an object
#### It is a method for an object. so, will show different nswers for each object

#### 2. Class Method : It is the method of the whole class or the full blueprint which will remain the same for all of the objects, then such kind of method is called as class method
#### In class method, these methods work with the class level attributes
#### They are going to use cls keyword instead of self keyword
#### They are called using the class name as well as the objects
#### For specifying that the current method is a class method we have to use the abbreviation as @classmethod

#### 3. Static Method --->  Static methods don't use self or cls keywords
#### They work like normal functions but belong to a class
#### They are going to be designated by an abbreviation @staticmethood
#### static method can also be called by the class name as well as object name also

```python
class Avengers :

    def __init__(self, name, power) :
        self.name = name
        self.power = power

    def show_details(self):
        print(f"My name is {self.name} and I am having power {self.power}")
```

```python
class Avengerss :

    origin = "Avenger"

    def __init__(self, name, power):
        self.name = name
        self.power = power

    @classmethod
    def origin_info(cls):
        print(f"My origin is {cls.origin}")

    def self_info(self):
        print(f"My name is {self.name} and power is {self.power}")
```

```python
Iron_man = Avengerss("Iron Man", "Genius Intellect")
```

```python
Iron_man.power
```

```python
Iron_man.self_info()
```

```python
Iron_man.origin_info()
```

```python
Avengerss.origin_info()
```

```python
Thor = Avengerss("Thor", "God of Thunder")
```

```python
Thor.self_info()
```

```python
Thor.origin_info()
```

```python
class Avengersss :

    @staticmethod
    def mission():
        print("All Avengers must protect the earth") # no attributes

    origin = "Avengers"

    def __init__(self,name,power):
        self.name = name
        self.power = power

    @classmethod
    def origin_info(cls):
        print(f"The origin for the Avenger is {cls.origin}")

    def introduction(self):
        print(f"My name is {self.name} and power is that i have {self.power}")
```

```python
iron_man = Avengersss("Iron Man", "Genius Intellect")
```

```python
iron_man.introduction() # instance method
```

```python
iron_man.origin_info() # class method
```

```python
iron_man.mission()
```

```python
Avengersss.mission()
```
