# Encapsulation

### Encapsulation is a way of wrapping data( variables, methods) inside a capsule or a class as a single unit

### It restricts the direct access to some of the objects component's which is a means of preventing accidental inferences and misuse of the data.

## Key concepts of Encapsulation
    
    1. Public Members
    2. Protected Members
    3. Private Members
    4. getter and setter method

### Public Members : ---> Accessible from anywhere or attributes are accessible outside of the class
### and these members are accessible by the derived class

### public member attribute is directly declared 

```python
class Ironman:

    def __init__(self, name, power, weapon):
        self.name = name          ## these are public members(attributes)
        self.power = power
        self.weapon = weapon

    def introduction(self):
        print(f"Hello, My name is {self.name} and my weapon is {self.weapon}")
```

#### dir() --> it takes the input ass the object name and gives us this idea that which member is public, private or protected

```python
iron = Ironman("Amit", "Genius Intellect", "Repulsor beams")
```

```python
dir(iron)
```

##### It shows up a list of attributes associated to the object : iron 

```python
iron.name
```

```python
iron.power
```

```python
iron.introduction()
```

###  Private Members : ---> A privte member is not accessible outside of the class and not even by the derived class or the child class. In order to have access over these private members or to bring some modifications to those members, we use getter and setter method

#### private members are declared using the double underscore _ before the name

```python
class Avenger:

    def __init__(self, name, phone_number):
        self.name = name
        self.__phone_number = phone_number
```

```python
avenge = Avenger("captain",48969216)
```

```python
avenge.phone_number
```

```python
avenge.__phone_number
```

#### we'll see public members are shown as it is
#### private members are shown asssociated with the class. meaning we can access or modify it from within the class

```python
class Captain:

    def __init__(self, name, address):
        self.name = name
        self.__address = address

    ## getter method

    def accessing_add(self):
        print(f"My address is {self.__address}")
```

```python
captain = Captain("Captain America", "New York")
```

```python
captain.address
```

```python
dir(captain)
```

```python
captain.__address
```

```python
captain.accessing_add()  # using another method to access the private data
```

```python
class Animal:

    def __init__(self, name, sound):
        self.name = name
        self.__sound = sound

    def sound_info(self):
        print(f"The animal is going to sound like :", self.__sound)

    def sound_modify(self, new_sound):
        self.__sound = new_sound
        print(f"The new sound of the animal is {self.__sound}")


class Dog(Animal):

    def __init__(self, name, sound):
        super().__init__(name,sound)

    def sound_information(self):
        print(f"The animal is going to sound like :", self.__sound)
```

```python
dog = Dog("bolt","woof")
print(dog.name)
dog.sound_info()
dog.sound_information()  # we can't access the private info from the derived getter method
```

### Protected Data is the data which is not accessible outside the class But the protected data is going to be accesible by the derived class or the child class

### for having the protected member you can mention your protected member as _attribute or by using a single underscore

```python
class Grandparent:

    def __init__(self, name, locker_key):
        self.name=name
        self._locker_key = locker_key

    # getter method
    def locker_info(self):
        print(f"The locker key info is {self._locker_key}")
```

```python
class Parent(Grandparent):

    def __init__(self, name, locker_key):
        super().__init__(name, locker_key)

    # getter method

    def child_reveals(self):
        print(f"The locker key information is : {self._locker_key}")
```

```python
grandpa = Grandparent("Amit", "123456")
```

```python
grandpa.name
```

```python
dir(grandpa)
```

```python
grandpa.locker_key
```

```python
grandpa._locker_key
```

```python
grandpa.locker_info()
```

```python
parent = Parent("Amit", "123456")
```

```python
parent.name
```

```python
parent.locker_key
```

```python
parent._locker_key
```

```python
parent.locker_info()
```

```python
parent.child_reveals()
```

```python
class BankAccount:
    def __init__(self,account_holder, balance):
        self.account_holder = account_holder
        self.__balance = balance

    ## getter
    def get_balance(self):
        print(f"The total balance is {self.__balance}")

    ## setter
    def deposit(self, amount):
        if amount>0:
            self.__balance+=amount
            self.get_balance()
        else:
            print("deposited amount must be positive")

    def withdraw(self, amount):
        if amount>0 and amount<=self.__balance :
            self.__balance-=amount
            self.get_balance()
        elif amount<0:
            print("withdrawl amount must be positive")
        else:
            print("Insufficient Balance")
            
```

```python
account = BankAccount("Amit Singh", 100000)
```

```python
account.balance
```

```python
account.get_balance()
```

```python
account.withdraw(5000)
```

```python
account.get_balance()
```
