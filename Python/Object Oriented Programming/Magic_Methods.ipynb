# Magic Methods

## Magic methods are also called as dunder methods because they have double underscores __ before and after their name.

## Magic Methods are special methods allowing us to customise how objects behave

### These methods can help us define :

    1. Object Initialisation (__init__)
    2. String Representation (__str__, __repr__)
    3. Arithmetic Operations (__add__, __sub__, etc)
    4. Comparison Operations (__eq__, __lt__)               equal to, lower than
    5. length, indexing, iteration (__len__, __getitem__, etc)

### Why to use Magic Methods 

    1. Makes objects behave like built in types (eg lists, integers)
    2. It improves readability by allowing basic operations like object1 + object2
    3. simplifies code by replacing verbose function calls with natural syntax

### print(object)  ---> __str__ ---> what is returned when print(object) is used

### arithmetic opertions can be done on objects using magic methods

```python
## __add__

class Avenger:

    def __init__(self, name, strength):
        self.name = name
        self.strength = strength

    def __add__(self,other):
        return Avenger(self.name + "&" + other.name, self.strength + other.strength)

    def __str__(self):
        return f"{self.name} with combined strength {self.strength}"


ironman =  Avenger("Iron Man", 80)
thor = Avenger("Thor", 100)

team = ironman + thor

print(ironman)
print(thor)
print(team)
```
