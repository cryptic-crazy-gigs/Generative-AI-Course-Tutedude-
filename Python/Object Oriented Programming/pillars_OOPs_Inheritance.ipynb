# Pillars of Object Oriented Programming

## Inheritance

#### Inheritance is a fundamental concept in OOPs because if which the child class is able to inherit the properties or Attributes and functionalities of the parent class

#### just like a child is going to inherit some properties fromm parent and parent is going to inherit the properties from the grand parents and definitely in this hierarchy the grandchild can have some features different from the parents as well as the grandparents


#### In python typically we get to see 3 different types of Inheritance :

1. Single Inheritance
2. Multilevel Inheritance
3. Multiple Inheritance

#### Single Inheritance ---> The features or the attributes from one parent class will be getting inherited into one child class.

##### Parent Class (properties : Attributes, functionalities)
##### |
##### |
##### Child Class


#### Multilevel Inheritance ---> The properties or methods and attributes of the grandparent class can be inherited by the parent class and then the properties of the parent class will be inherited by the child class. In a proper hierarchy, the inheritance process is going to work for us. 

##### Grandparent (properties)
##### |
##### |
##### Parent (properties)
##### |
##### |
##### Child


#### Multiple Inheritance ---> In multiple Inheritance, The properties of two parent class or multiple parent class will be getting inherited into the child class 

##### Parent 1 (properties) + Parent 2 (properties) + Parent 3 (properties) + ...
##### |
##### |
##### Child

```python
class Avenger :

    def __init__(self, name, power):
        self.name = name
        self.power = power

    def introduction(self):
        print(f"Hello my name is {self.name}")

    
class Ironman(Avenger) :

    def __init__(self,name,power,weapon):
        super().__init__(name,power)  # super() to inherit the attributes from the parent class
        self.weapon = weapon

    def weapon_details(self):
        print(f"I use {self.weapon}")
```

```python
iron = Ironman("Iron Man", "Genius Intellect", "Repulsor Beams")
```

```python
iron.name  # name -> parent attribute
```

```python
iron.weapon  # weapon -> child attribute
```

```python
iron.weapon_details()  # weapon_details -> child functionality
```

```python
iron.introduction()  # introduction -> parent functionality
```

```python
class Grandparent:
    
    def __init__(self, nationality, baldness, anger_issues):
        self.nationality = nationality
        self.baldness = baldness
        self.anger_issues = anger_issues

    def introduction(self):
        print(f"Hello my nationality is : {self.nationality} and about my anger issues : {self.anger_issues}") 
```

```python
class Parent(Grandparent):

    def __init__(self, nationality, baldness, anger_issues, intelligence):
        super().__init__(nationality, baldness, anger_issues)
        self.intelligence = intelligence

    def intelligence_details(self):
        print(f"About my intelligence, if you want to know, it is {self.intelligence}")
```

```python
class Child(Parent):

    def __init__(self, nationality, baldness, anger_issues, intelligence, charm):
        super().__init__(nationality, baldness, anger_issues, intelligence)
        self.charm = charm

    def charm_info(self):
        print(f"How much charm i have : {self.charm}")
```

```python
parent_1 = Parent("Indian", "Yes", "No", "Yes")
```

```python
parent_1.introduction()
```

```python
child_1 = Child("Indian", "No", "Yes", "Yes", "Yes")
```

```python
child_1.introduction()
```

```python
child_1.intelligence_details()
```

```python
child_1.charm_info()
```

```python
child_1.nationality = "Christian"
```

```python
child_1.introduction()
```

```python
class Ironman :

    def __init__(self, name, power, suit, weapon, nationality = "American"):
        self.name = name
        self.power = power
        self.weapon = weapon
        self.nationality = nationality

    def introduction(self):
        print(f"My name is {self.name} and powers are : {self.power}")


class Captainamerica :

    def __init__(self, name, power, suit, weapon, nationality = "American"):
        self.name = name
        self.power = power
        self.weapon = weapon
        self.nationality = nationality

    def weapon_details(self):
        print(f"I carry the weapon {self.weapon}")
    
```

```python
class Hulk(Ironman, Captainamerica) :

    def __init__(self, name, power, suit, weapon, nationality = "American", anger_issues):
        super().__init__(name, power, suit, weapon, nationality)
        self.anger_issues = anger_issues

    def anger_description(self):
        print(f"My anger issues : {self.anger_issues}")
```

```python
hulk_guy = Hulk("Hulk", "Amazing Power and Strength", "Mark 85 Ironman suit", "Repulsor beams and Shield", "American", "Yes")
```

```python
hulk_guy.introduction()
```

```python
hulk_guy.weapon_details()
```

```python
hulk_guy.anger_description()
```
