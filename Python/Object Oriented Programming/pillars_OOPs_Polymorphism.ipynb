# Polymorphism

### Polymorphism means MANY FORMS in greek

### In OOPs, polymorphism allows objects of different classes as if they are the objects of the same class and it enables reusability and makes programs more flexible

### Types of Polymorphism

    1. Method Overriding (runtime polymorphism)
    2. Method Overloading (compile time polymorphism)
    3. Operator Overloading


Method Overriding  ---- > A child class provides a specific implementation of a method that has already been defined into the parent class. This allows dynamic selection at runtime. 

```python
class Grandparent :    # Parent Class

    def speak(self):
        print("Hello My name is Amit")

class Parent (Grandparent) :  # Child 1 class
    
    def speak(self):
        print("Hello My name is Junior Amit")

class Child (Grandparent) :   # Child 2 class

    def speak(self):
        print("Hello My name is Junior Junior Amit")
```

```python
def attribute_provider(objects):
    objects.speak()
```

```python
grandpa = Grandparent()
```

```python
attribute_provider(grandpa)
```

```python
parent = Parent()
```

```python
attribute_provider(parent)
```

```python
child = Child()
```

```python
attribute_provider(child)
```

```python
class Shape:

    def area(self):
        return "This method returns the area of the shape"
```

```python
class Rectangle(Shape):

    def __init__(self,width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width*self.height
```

```python
class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14*self.radius*self.radius
```

child class having a different implemetnation of the same method : area()

```python
shape = Shape()
print(shape.area())
```

```python
rect = Rectangle(5,7)
print(rect.area())
```

```python
circle = Circle(5)
print(circle.area())
```

#### shape -> parent class   : area method
#### rect, circle -> child class : different implementations of area method

#### rect and circle classes having different forms of the same area method : Polymorhism : MANY FORMS

Virtual functions are used to achieve runtime polymorphism. They allow a derived class to override a base class method, and when called through a base class reference or pointer, the method of the actual object type is executed at runtime. In Python, methods are virtual by default, so there is no explicit virtual keyword.
