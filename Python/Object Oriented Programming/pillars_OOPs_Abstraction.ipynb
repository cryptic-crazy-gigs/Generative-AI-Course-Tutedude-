# Abstraction

### Abstraction is the concept of hiding the implementation details and only showing the necessary details/ features about the object
### this helps in reducing programming complexity and efforts

### Abstraction allows us to work with an object with what an object does rather than how it does it

### How can we achieve abstraction in python ?

#### 1. Abstract class : A class that can't be instantiated or that means for an abstract class, we can't create the object of this class
#### 2. Abstract method : Abstract method just because it can't be accessed by the object of the abstract class itself that's why we have to call this abstract method in our child class and the implementation of this abstract method is not provided inside the abstract class but must be provided in the child class
#### the abstract method is being defined by @abstractmethod     

abstract method is the one with nothing for implementation and abreviation @abstractclass

abstract class is the one wchich satisfies both:
1. it is a child to ABC
2. has at;east one abstract method.
one of above not fulfilled -> not an abstract class

```python
from abc import ABC,abstractmethod

class Vehicle(ABC):
    
    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass

class Car(Vehicle):

    def start(self):
        print("The car starts with an ignition key")

    def stop(self):
        print("The car stops with break pedals")


class ElectricScooter(Vehicle):

    def start(self):
        print("Electric Scooter is going to start with a power button")

    def stop(self):
        print("Electric Scooter stops with an electric break")
```

```python
car = Car()
```

```python
car.start()
```

```python
scooter = ElectricScooter()
```

```python
scooter.start()
```

```python
scooter.stop()
```

```python
class PaymentMethod(ABC):

    @abstractmethod
    def pay(self, amount):
        pass
```

```python
class CreditCard(PaymentMethod):
    def pay(self, amount):
        print(f"paid {amount} using Credit Card")

class UPI(PaymentMethod):
    def pay(self, amount):
        print(f"paid {amount} using UPI")

class PayPal(PaymentMethod):
    def pay(self, amount):
        print(f"paid {amount} using PayPal")
```

```python
cc = CreditCard()
upi = UPI()
paypal = PayPal()
```

```python
def checkout(payment_method, amount):
    payment_method.pay(amount)
```

```python
checkout(cc, 500)
```

```python
checkout(paypal, 100)
```

```python
checkout(upi, 1408)
```
