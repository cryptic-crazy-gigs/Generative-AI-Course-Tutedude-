```python
## we are cretaing a function to see that how the translation is going to be there
# from english to hindi language

def translator(english_text):
    """ This function is going to convert english language to hindi language"""

    if english_text == "Hello" or english_text=="hello" :
        print(f"translation of {english_text} is namaste")
    elif english_text == "help" or english_text == "Help":
        print(f"translation of {english_text} is madad")
    elif english_text == "required" or english_text=="Required":
        print(f"translation of {english_text} is aavashyak")
```

for having access to our function and to provide input to our function, we simply call our function

```python
# calling a function --> Mention the name of the function and then provide the input in ()

english_input = input("Enter your english word for getting it translated into Hindi")

translator(english_input)
```

```python
def even_or_odd_value(num):
    if type(num)==str:
        print("Invalid Input Type")
    elif num%2==0 :
        print("Even Number")
    else :
        print("Odd Number")
```

```python
even_or_odd_value("s")
```

let's do some error handling another way.
using isinstance function


```python
def even_or_odd_instance(num):
    if isinstance(num, int):
        if num%2==0:
            print("even number")
        else :
            print("odd number")
    elif isinstance(num, str):
        print("Invalid Input Type")
    else :
        print("Invalid")
```

```python
even_or_odd_instance(5)
```

```python
even_or_odd_instance(10)
```

```python
even_or_odd_instance("s")
```

```python
even_or_odd_instance(7.3)
```
