```python
## csv library -----> Most of the time we are going to work with csv files so thats why we need to have the knowledge of 
# csv library functionalities
```

```python
# 1. creating the csv file. 
# 2. writing content into it.
# 3. read the contents of the file
```

```python
import csv
```

```python
with open('example.txt', mode = 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Name','Age'])
    writer.writerow(['Amit', 23])
```

```python
with open('example.txt', mode ='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)

        # row is a list
        # reader is an iterable which has got list as elements
```

```python
type(reader)
```

```python
## when we are going to work with Nosql Database Data then the data inside the NoSQL database is going to be their in the json 
# format and along with that when we are going to work with the datasets as well then again thse dataset 
# will be consisting in json format itself
```

```python
# when wwe get the json data, it is in json string format. So by using json library we can convert this string 
# into dictionary format for serializing the number

# The whole process is called serializing
```

```python
import json
```

```python
data = {'Name':'Amit', 'Age':25}
json_str = json.dumps(data)
print(json_str)
print(type(json_str))  

# json.dumps  ddumps dictionary to json string format
```

```python
json_dict = json.loads(json_str)
print(json_dict)
print(type(json_dict))
```

```python
# Regular Expression library

## regular expression is used to retrieve information based on a specific pattern
# or if the pattern matches
```

```python
import re
```

```python
text = 'My name is Amit and my password is HelloAmit2025'

pattern = r'\d+'
match = re.search(pattern, text)
print(match)
```

```python
text = 'My name is Amit and my password is HelloAmit2025'

pattern = r'\d+'
match = re.search(pattern, text)
print(match.group())
```

```python
text = 'My name is Amit and my password is HelloAmit2025'

pattern = r'\d'
match = re.search(pattern, text)
print(match.group())
```
