## Types of operators

1. Arithmetic operators

Addition (+)
sub (-)
multiplication(*)
division(/)
integer division(//)
modulus (%)
exponentiation(a**b)  a^b

2. Comparison Operators/ Relational Operators

greater than (>)
greater than or equal to(>=)
less than (<)
less than or equal to(<=)
is equal to (==)
is not equal to (!=)

3. Logical Operators

and -> if both conditions are true or not
or-> if one of the conditions are true or not
not -> negation or to inverse the given result
