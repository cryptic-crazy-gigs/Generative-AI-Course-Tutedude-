# conditional Statements

### if a certain condition is met, a code of block is executed

```python
Age = int(input("enter youor age please : "))

if Age>=18:
    print("You are eligible to vote")
else :
    print("You are ineligible to vote")
```

## conditional statements 
### if
### elif
### else

```python
# figure out who is tallest among 2 friends

amit = float(input("enter you height, Amit :"))
prabhat = float(input("enter your height, Prabhat :"))

if amit > prabhat :
    print("Amit is taller")
elif amit == prabhat:
    print("amit and prabhat have same height")
else:
    print ("prabhat is taller")
```

## Nested Conditional statements

### if else inside if else



```python
# figure who is tallest among 3
amit = float(input("enter you height, Amit :"))
prabhat = float(input("enter your height, Prabhat :"))
shivy = float(input("enter your height, Prabhat :"))

if amit > prabhat :
    if amit > shivy :
        print("amit is tallest")
    else :
        print("shivy is tallest")
else:
    if prabhat > shivy :
        print("prabhat is tallest")
    else:
        print("shivy is tallest")

```

```python
### whether current year is a leap year or not

current_year = int(input("enter current year :"))

if current_year % 4 == 0 :
    if current_year % 100 == 0:
        if current_year % 400 == 0:
            print("Leap year")
        else :
            print("Not a leap year")
    else :
        print("Leap year")
else :
    print("Not a leap year")
```
